```
from google.colab import files
uploaded = files.upload()
```

```
import io
import pandas as pd
import numpy as np
df = pd.read_csv(io.BytesIO(uploaded['movie_metadata.csv']))
#df.pop('description')
df.head()
```

```
df = df.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title','imdb_score']]
df
```

```
print(df['actor_1_name'].isnull().values.any())
print(df['director_name'].isnull().values.any())
print(df['actor_2_name'].isnull().values.any())
print(df['actor_3_name'].isnull().values.any())
print(df['genres'].isnull().values.any())
print(df['movie_title'].isnull().values.any())
print(df['imdb_score'].isnull().values.any())
```

```
df['actor_1_name'] = df['actor_1_name'].replace(np.nan, 'unknown')
df['actor_2_name'] = df['actor_2_name'].replace(np.nan, 'unknown')
df['actor_3_name'] = df['actor_3_name'].replace(np.nan, 'unknown')
df['director_name'] = df['director_name'].replace(np.nan, 'unknown')
```

```
print(df['actor_1_name'].isnull().values.any())
print(df['director_name'].isnull().values.any())
print(df['actor_2_name'].isnull().values.any())
print(df['actor_3_name'].isnull().values.any())
print(df['genres'].isnull().values.any())
print(df['movie_title'].isnull().values.any())
print(df['imdb_score'].isnull().values.any())
```

```
df['genres'] = df['genres'].str.replace('|', ' ')
```

```
df.head()
```

```
df['movie_title'] = df['movie_title'].str.lower()
df['movie_title'][1]
```

```
df['movie_title'] = df['movie_title'].str.replace('\xa0', '')
df['movie_title'][1]
```

```
df=df.sort_values('imdb_score',ascending=False)
```

```
df['combine'] = df['director_name'] +" "+df['actor_1_name'] +" "+df['actor_2_name'] +" "+df['actor_3_name'] +" "+df['genres'] +" "+df['movie_title'] 
df['combine']
```
